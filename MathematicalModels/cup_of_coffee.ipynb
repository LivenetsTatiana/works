```python
import matplotlib.pyplot as plt
from google.colab import widgets
from google.colab import output
from IPython.display import Image, display

def coffee(temperature, temperature2, temperature3, time):
    T = T2 = T3 =temperature[0]
    t=0
    while (T>45):
        T = T - k*(T-Te)*dt
        temperature = temperature + [T]
        T2 = T2 - k*(T-Te)*dt
        temperature2 = temperature2 + [T2]
        T3 = T3 - k*((T+T2)/2-Te)*dt
        temperature3 = temperature3 + [T3]
        t+=dt
        time = time + [t]
    #print(T, t)
    fig = plt.figure(figsize=(10, 7), dpi=100)
    ax = fig.add_subplot()
    ax.plot(time, temperature, 'r')
    ax.plot(time, temperature2, 'b')
    ax.plot(time, temperature3, 'g')
    ax.grid()
    ax.set_ylabel('temperature')
    ax.set_xlabel('time')
    plt.title("Cup of coffee")
    plt.show()

def coffee_with_milk(temperature, temperature2, temperature3, time):
    T = T2 = T3 =temperature[0]
    t=0
    Tmilk = 22
    mwater = 0.15
    mmilk = 0.1

    while (T>45) and (t<2):
        T = T - k*(T-Te)*dt
        temperature = temperature + [T]
        T2 = T2 - k*(T-Te)*dt
        temperature2 = temperature2 + [T2]
        T3 = T3 - k*((T+T2)/2-Te)*dt
        temperature3 = temperature3 + [T3]
        t+=dt
        time = time + [t]
    #print(T, t)
    Tmix = (mwater*T+mmilk*Tmilk)/(mmilk+mwater)
    T = T2 = T3 =Tmix
    while (T>45):
        T = T - k*(T-Te)*dt
        temperature = temperature + [T]
        T2 = T2 - k*(T-Te)*dt
        temperature2 = temperature2 + [T2]
        T3 = T3 - k*((T+T2)/2-Te)*dt
        temperature3 = temperature3 + [T3]
        t+=dt
        time = time + [t]
    #print(T, t)
    #print(T2, t)
    #print(T3, t)
    fig = plt.figure(figsize=(10, 7), dpi=100)
    ax = fig.add_subplot()
    ax.plot(time, temperature, 'r')
    ax.plot(time, temperature2, 'b')
    ax.plot(time, temperature3, 'g')
    ax.grid()
    ax.set_ylabel('temperature')
    ax.set_xlabel('time')
    plt.title("Cup of coffee with milk")
    plt.show()

def coffee_with_milk_difference(temperature, temperature2, temperature3, time):
    T = T2 = T3 =temperature[0]
    t=0
    Tmilk = 22
    mwater = 0.15
    mmilk = 0.1

    while (T>45) and (t<2):
        T = T - k*(T-Te)*dt

        T2 = T2 - k*(T-Te)*dt
        temperature2 = temperature2 + [T2-T]
        T3 = T3 - k*((T+T2)/2-Te)*dt
        temperature3 = temperature3 + [T3-T]
        temperature = temperature + [T-T2]
        t+=dt
        time = time + [t]
    #print(T, t)
    Tmix = (mwater*T+mmilk*Tmilk)/(mmilk+mwater)
    T = T2 = T3 =Tmix
    while (T>45):
        T = T - k*(T-Te)*dt
        temperature = temperature + [T]
        T2 = T2 - k*(T-Te)*dt
        temperature2 = temperature2 + [T2]
        T3 = T3 - k*((T+T2)/2-Te)*dt
        temperature3 = temperature3 + [T3]
        t+=dt
        time = time + [t]
    #print(T, t)
    #print(T2, t)
    #print(T3, t)
    fig = plt.figure(figsize=(10, 7), dpi=100)
    ax = fig.add_subplot()
    ax.plot(time, temperature, 'r')
    ax.plot(time, temperature2, 'b')
    ax.plot(time, temperature3, 'g')
    ax.grid()
    ax.set_ylabel('temperature')
    ax.set_xlabel('time')
    plt.title("Cup of coffee with milk")
    plt.show()

T0=83
Te=22
T1=77.7
dt=0.1
T=T0
k = 0.09
T2 = T
T3 = T
t=0
temperature = [T]
temperature2 = [T]
temperature3 = [T]
time = [t]

t = widgets.TabBar(["use_case", "state_chart", "coffee", "coffee_with_milk"])
with t.output_to(0):
  display(Image('cup_usecase.drawio.png'))
with t.output_to(1):
  display(Image('cup_stagechart.drawio.png'))
with t.output_to(2):
  coffee(temperature, temperature2, temperature3, time)
with t.output_to(3):
  coffee_with_milk(temperature, temperature2, temperature3, time)
with t.output_to(4):
  coffee_with_milk(temperature, temperature2, temperature3, time)
```
